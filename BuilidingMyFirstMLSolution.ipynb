Data Scaling - unit of measure is important, because it impacts the magnitude, therefore affecting the results<br>

```python
import matplotlib.pyplot as plt
plt.plot(160, 63,'rx')
plt.plot(170.2, 67,'gx')
plt.plot(177.8, 70,'bx')
plt.plot(150, 63,'ro')
plt.plot(160.2, 67,'go')
plt.plot(171.8, 70,'bo')
plt.show()
```

```python
import numpy as np
distance1 = np.linalg.norm(np.array((160, 63)) - np.array((170.2, 67)))
distance2 = np.linalg.norm(np.array((177.8, 70)) - np.array((170.2, 67)))
print(f'1-2: {distance1}, 1-3: {distance2}')
print('Lowest distance =', min(distance1, distance2))
```

## Training the model

```python
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
from sklearn.linear_model import LinearRegression, ElasticNet, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, KFold, cross_val_score
```

```python
#filename="forestfires.csv"
url="https://raw.githubusercontent.com/smarter-code/PS-BuildingYourFirstMachineLearningSolution/main/forestfires.csv"
names = ['X', 'Y', 'month', 'day', 'FFMC', 'DMC',
         'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']
df = pd.read_csv(url, names=names)
months = {'jan' : 1, 'feb' : 2, 'mar' : 3, 'apr' : 4,
          'may' : 5, 'jun' : 6, 'jul' : 7, 'aug' : 8,
          'sep' : 9, 'oct' : 10, 'nov' : 11, 'dec' : 12}
days = {'mon' : 1, 'tue' : 2, 'wed' : 3, 'thu': 4, 'fri' : 5, 'sat' : 6, 'sun': 7}
df.month.replace(months, inplace=True)
df.day.replace(days, inplace=True)
print(df.head())
```

```python
array = df.values
X = array[:,0:12] # first 12 columns
Y = array[:,12] # 12th column - what is to be predicted
```

```python
max_error_scoring = 'max_error'
neg_mean_absolute_error_scoring = 'neg_mean_absolute_error'
r2_scoring = 'r2'
neg_mean_squared_error_scoring = 'neg_mean_squared_error'
```

```python
models = []
models.append(('LR', LinearRegression()))
models.append(('LASSO', Lasso()))
models.append(('EN', ElasticNet()))
models.append(('Ridge', Ridge()))
models.append(('KNN', KNeighborsRegressor()))
models.append(('SVR', SVR()))
models
```

```python
# Evaluate models and print results
results = []
names = []
for name, model in models:
  kfold = KFold(n_splits=10, random_state=7)
  cv_results1 = cross_val_score(model, X, Y, cv=kfold, scoring=max_error_scoring)
  cv_results2 = cross_val_score(model, X, Y, cv=kfold, scoring=neg_mean_absolute_error_scoring)
  cv_results3 = cross_val_score(model, X, Y, cv=kfold, scoring=r2_scoring)
  cv_results4 = cross_val_score(model, X, Y, cv=kfold, scoring=neg_mean_squared_error_scoring)
  print(f"{name}:\nmax error: {cv_results1.mean()}\nmean absolute error: {-cv_results2.mean()}\nr2: {cv_results3.mean()}\nmean squared error: {-cv_results4.mean()}\n")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=1, shuffle=True)
```

```python
lasso_model = Lasso()
lasso_model.fit(X_train, Y_train) # fit trains the ML model i.e. we fit the training data to our model
```

```python
predictions = lasso_model.predict(X_test)
len(predictions)
```

```python
expected = Y_test
len(expected)
```

```python
import numpy as np
print("RMS: %r " % np.sqrt(np.mean((predictions - expected) ** 2)))
```

```python
plt.figure(figsize=(7,5))
plt.scatter(Y_test, predictions)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r-', lw=2)
plt.xlabel('Expected')
plt.ylabel('Predicted')
plt.show()
```
